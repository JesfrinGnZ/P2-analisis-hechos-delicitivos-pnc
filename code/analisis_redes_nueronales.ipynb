# Exportar librerías

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential #neuronas
from tensorflow.keras.layers import Dense #conexiones
```

# Carga de datos

```python
detenidos_data = pd.read_csv('detenidos.csv')
detenidos_data.head()
```

```python
victimas_data = pd.read_csv('victimas.csv')
victimas_data.head()
```

# Modelo 1

Predicción de edad de la víctima en base al delito que le cometieron

```python

from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split


# Filtrado
victimasFpg_f2 = (
    victimas_data[victimas_data['edad_per'].isin([15, 16, 17])]
    .drop(columns=['menor_mayor'])
)

# Aleatoriedad
victimasFpg_f2 = victimasFpg_f2.sample(frac=1, random_state=42).reset_index(drop=True)

# Definición de variables independientes y dependientes
x = victimasFpg_f2[['delito_com']]
y = victimasFpg_f2[['edad_per']]

# One hot encoding
encoder = OneHotEncoder(sparse_output=False)
y_encoded = encoder.fit_transform(y)

# Division de datos en entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.2, random_state=42)

# Creacion d emodelo
model = Sequential()
model.add(Dense(8, input_dim=x_train.shape[1], activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(3, activation='softmax'))

# Compilacion del modelo
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Entrenando modelo
model.fit(x_train, y_train, epochs=50, batch_size=8, validation_split=0.2)

# Evaluacion del modelo
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Pérdida: {loss}, Exactitud: {accuracy}')
```

```python
#predicciones
# Lista de delitos
delitos = {
    1: "Homicidio por arma de fuego",
    2: "Homicidio por arma blanca",
    15: "Secuestro",
    20: "Robo a comercios",
    22: "Robo a vehículos",
    33: "Extorsión a comercios"
}


for delito_com, descripcion in delitos.items():
    entrada = np.array([[delito_com]])
    p = model.predict(entrada)
    probabilidades = p[0] * 100

    # resultados
    print(f"Probabilidades dado que el delito es {descripcion}:")
    for edad, probabilidad in zip([15, 16, 17], probabilidades):
        print(f" - {edad}: {probabilidad:.2f}%")
    print()
```

# Modelo 2

Predicción del grupo de hora en la que fue deteneida una persona en base al delito que cometió.

```python

from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split

detenidos_f1 = detenidos_data;

# Aleatoriedad
detenidos_f1 = detenidos_f1.sample(frac=1, random_state=42).reset_index(drop=True)

# Definición de variables independientes y dependientes
x = detenidos_f1[['delito_com']]
y = detenidos_f1[['g_hora']]

# Codificar la variable dependiente (y) en One-Hot Encoding
encoder = OneHotEncoder(sparse_output=False)
y_encoded = encoder.fit_transform(y)

# División de datos en entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.2, random_state=42)

# Crear el modelo de red neuronal
model = Sequential()
model.add(Dense(16, input_dim=x_train.shape[1], activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(y_train.shape[1], activation='softmax'))

# Compilar el modelo
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
model.fit(x_train, y_train, epochs=10, batch_size=8, validation_split=0.2)

# Evaluar el modelo en el conjunto de prueba
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Pérdida: {loss}, Precisión: {accuracy}')
```

```python
# grupos de hora
grupo_hora_descripcion = {
    1: "00:00 a 05:59",
    2: "06:00 a 11:59",
    3: "12:00 a 17:59",
    4: "18:00 a 23:59",
    5: "Ignorada"
}

# Predicciones
gh1 = np.array([[1]])  # Homicidio por arma de fuego
p = model.predict(gh1)[0]


print("Grupo de hora de asalto a persona por homicidio por arma de fuego:")
for i, probabilidad in enumerate(p, start=1):
    print(f" - Grupo {i} ({grupo_hora_descripcion[i]}): {probabilidad * 100:.2f}%")

# Predicciones
gh2 = np.array([[48]])  # Riña
p = model.predict(gh2)[0]

print("\nGrupo de hora de detención por Riña:")
for i, probabilidad in enumerate(p, start=1):
    print(f" - Grupo {i} ({grupo_hora_descripcion[i]}): {probabilidad * 100:.2f}%")
```
